## Standardization
---

```python
url = 'https://drive.google.com/file/d/1ERGkh-O_Zd34u4kJvcMfCakY8wKmZMcd/view?usp=sharing'
SP500 = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
```

```python
sp500=pd.read_csv(SP500,index_col=0)
sp500.head()
```

```python
sp500["return"]=sp500["Close"].pct_change() # Percentage change
```

### Max-min standardization

```python

sp500["max_min"]=(sp500["return"]-sp500["return"].min())/(sp500["return"].max()-sp500["return"].min())
sp500.head()
```

```python
plt.hist(sp500["return"],bins=30,edgecolor="black",density=True)
plt.show()
```

```python
plt.hist(sp500["max_min"],bins=30,edgecolor="black",density=True)
plt.show()
```

```python
sp500["mean_sd"]=(sp500["return"]-sp500["return"].mean())/sp500["return"].std(ddof=1)
sp500.head()
```

```python
plt.hist(sp500["return"],bins=30,edgecolor="black",density=True)
plt.show()
```

```python
plt.hist(sp500["mean_sd"],bins=30,edgecolor="black",density=True)
plt.show()
```

## Rolling and Cumulative Methods
---

```python
dates = pd.date_range('20210105', periods=20, freq='D') # using pandas method date_range to make a index
dates
```

```python
stock=pd.DataFrame(np.random.randn(20),index=dates,columns=["daily return"])
stock.head()
```

```python
stock["RollMean5"]=stock["daily return"].rolling(5).mean()
stock.head(8)
```

```python
stock["RollSum5"]=stock["daily return"].rolling(5).sum()
stock.head(8)
```

```python
stock["CumSum"]=stock["daily return"].cumsum()
stock.head(8)
```
