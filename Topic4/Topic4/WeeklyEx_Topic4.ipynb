```python
# Import the data
import pandas as pd

import numpy as np

# Graph plotting
import matplotlib.pyplot as plt
%matplotlib inline

# import the function StandardScaler
from sklearn.preprocessing import StandardScaler

# hierarchical clustering
from scipy.cluster.hierarchy import dendrogram, linkage #package for drawing the dendrogram and define linkage
from sklearn.cluster import AgglomerativeClustering

# Kmeans clustering
from sklearn.cluster import KMeans

# Disable warnings
import warnings
warnings.filterwarnings('ignore')

from scipy import stats
```

```python
def Dendrogram(X, method, metric):
    plt.figure(figsize=(10, 7))  
    Z = linkage(X, method=method, metric=metric)
    dendrogram(Z, labels = range(1, len(X) + 1))
    plt.title("Dendogram") 
    plt.xlabel('ID')
    plt.ylabel('distiance')
    plt.show()
```

```python
x1=[7,15,11,5,8]
x2=[14,16,16,14,24]
```

```python
plt.scatter(x1,x2)
plt.xlabel('x1',fontsize=12)
plt.ylabel('x2',fontsize=12)

for i in range(0,5):
    plt.annotate(i+1, (x1[i]+0.05, x2[i]))
```

```python
X=np.array([x1,x2]).T
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
Z = linkage(X, method = 'complete', metric = 'euclidean')
dendrogram(Z, labels = range(1, len(X) + 1))
plt.title('Dendrogram-1')

X=X.sum(axis=1).reshape(5,1)
plt.subplot(2,2,2)
Z = linkage(X, method = 'complete', metric = 'euclidean')
dendrogram(Z, labels = range(1, len(X) + 1))
plt.title('Dendrogram-2')

X=np.array([x1]).T
plt.subplot(2,2,3)
Z = linkage(X, method = 'single', metric = 'euclidean')
dendrogram(Z, labels = range(1, len(X) + 1))
plt.title('Dendrogram-3')

plt.subplot(2,2,4)
Z = linkage(X, method = 'centroid', metric = 'euclidean')
dendrogram(Z, labels = range(1, len(X) + 1))
plt.title('Dendrogram-4')


plt.show()
```
